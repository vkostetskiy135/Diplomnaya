## Обзор фреймворков Django, Flask и FastAPI
### Django
**Django** — это высокоуровневый фреймворк для веб-разработки на Python, созданный с целью ускорения создания сложных и полнофункциональных веб-приложений. Django придерживается философии "всеобъемлющего" подхода, предлагая множество встроенных инструментов, таких как ORM (Object-Relational Mapping), система аутентификации пользователей, панель администрирования и маршрутизация URL. Это делает Django идеальным выбором для больших и сложных проектов, где требуется наличие всех необходимых компонентов "из коробки".

### Особенности Django:

- Быстрая разработка благодаря встроенным инструментам и библиотекам.
- Безопасность: встроенные механизмы защиты от большинства распространенных веб-угроз.
- Поддержка множества сторонних приложений, расширяющих возможности фреймворка.
- Идеален для создания многофункциональных корпоративных приложений.

### Flask
**Flask** — это микрофреймворк для веб-разработки, известный своей простотой и гибкостью. В отличие от Django, Flask предоставляет только базовые инструменты для разработки, оставляя возможность выбора дополнительных библиотек и компонентов. Этот фреймворк особенно популярен среди разработчиков, которые предпочитают минимализм и контроль над каждым аспектом приложения.

### Особенности Flask:

- Минималистичный подход: только основные компоненты без лишних зависимостей.
- Высокая гибкость, позволяющая легко интегрировать сторонние библиотеки и модули.
- Поддержка расширений, таких как Flask-SQLAlchemy для работы с базами данных и Flask-WTF для форм.
- Подходит для небольших и средних проектов, где важна гибкость и легковесность.

### FastAPI
**FastAPI** — это современный, высокопроизводительный фреймворк для создания веб-приложений и API, построенный на стандартах OpenAPI и JSON Schema. Он использует асинхронное программирование (async/await), что позволяет достигать высокой производительности и масштабируемости. FastAPI также автоматически генерирует интерактивную документацию API, что делает его удобным для разработчиков.

###Особенности FastAPI:

- Высокая производительность благодаря асинхронному программированию.
- Поддержка валидации данных и аннотаций типов данных на уровне системы.
- Автоматическая генерация документации API с помощью Swagger UI и ReDoc.

## Реализация веб-приложений

В Django для реализации данного приложения есть возможность использовать встроенное средство для работы с пользователями и сессиями. Django предоставляет встроенную аутентификацию и обработку форм, что значительно ускорило процесс разработки. ##Flask
В проекте на Flask было создано простое веб-приложение для регистрации и авторизации пользователей. Все ключевые компоненты, такие как работа с пользователями и сессиями, были реализованы вручную, без использования расширений вроде Flask-Login или Flask-WTF.

## FastAPI
В отличие от других фреймворков, FastAPI изначально ориентирован на создание высокопроизводительных API, но также может быть использован для разработки стандартных веб-приложений. Валидация данных происходила на уровне маршрутов с использованием аннотаций типов и Pydantic.

###Ключевые моменты реализации:

2. Аутентификация: В приложении использовалась стандартная авторизация с сохранением сессий.
